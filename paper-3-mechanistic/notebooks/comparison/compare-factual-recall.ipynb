# Cross-Model Comparison: Factual Recall

Loads saved results from GPT-2 Medium, Large, and XL baseline notebooks.
Compares when and how each model size retrieves factual knowledge.

**Run the individual model baselines first** — this notebook only loads CSVs.

```python
import sys
sys.path.insert(0, '../..')

import pandas as pd
from src.viz import plot_phase_comparison
```

## France: When does each model find Paris?

```python
# Load logit lens results
models_france = {}
for model_name in ["gpt2-medium", "gpt2-large", "gpt2-xl"]:
    try:
        df = pd.read_csv(f"../results/{model_name}/paris-logit-lens.csv")
        models_france[model_name] = df
        print(f"Loaded {model_name}")
    except FileNotFoundError:
        print(f"Not yet: {model_name} (run its baseline notebook first)")
```

```python
if models_france:
    plot_phase_comparison(models_france, metric="target_rank")
    plot_phase_comparison(models_france, metric="target_prob")
```

## Head Comparison

Do the same heads fire across model sizes? Load head CSVs and compare
top contributors.

```python
for model_name in ["gpt2-medium", "gpt2-large", "gpt2-xl"]:
    try:
        df = pd.read_csv(f"../results/{model_name}/paris-heads.csv")
        top = df.nlargest(5, "abs_logit")[["layer", "head", "logit"]]
        print(f"\n{model_name} — Top 5 heads for Paris:")
        print(top.to_string(index=False))
    except FileNotFoundError:
        print(f"\nNot yet: {model_name}")
```
